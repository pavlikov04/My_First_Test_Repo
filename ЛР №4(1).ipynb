```python
import numpy as np
import matplotlib.pyplot as plt

t=0.01
x1=np.arange(-2,0,t)
y1=-x1
x2=np.arange(0,np.pi,t)
y2=np.sin(x2)
x3=np.arange(np.pi,2*np.pi,t)
y3=0*x3
x4=np.arange(2*np.pi,3*np.pi,t)
y4=np.sin(x4)
x5=np.arange(3*np.pi,11,t)
y5=x5-3*np.pi
plt.plot(x1,y1,lw=2,label='1')
plt.plot(x2,y2,lw=2,label='2')
plt.plot(x3,y3,lw=2,label='3')
plt.plot(x4,y4,lw=2,label='4')
plt.plot(x5,y5,lw=2,label='5')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('График')
plt.legend()





```
